```python
import pandas as pd
df = pd.read_excel('EntriesGender.xlsx')
df = df.sort_values('Total',ascending=False)
df
```

```python
male = (
    df.loc[:, ['Discipline', 'Male']]

)
male_age = male['Discipline'].to_list()  # 项目
male_people = male['Male'].to_list()  # 人数
```

```python
female = (
    df.loc[:, ['Discipline', 'Female']]

)
female_age = female['Discipline'].to_list()  # 项目
female_people = female['Female'].to_list()  # 人数
```

```python
male_val = []
for a, m in zip(male_age, male_people):
    male_val.append({"year": 2022, "age": a, "sex": 1, "people": m})
male_val
```

```python
female_val = []
for a, f in zip(female_age, female_people):
    female_val.append({"year": 2022, "age": a, "sex": 2, "people": f})
female_val
```

```python
val = female_val + male_val
val
```

```python
import json
with open('数据.json', 'w') as f:
    f.write(json.dumps(val))
```
