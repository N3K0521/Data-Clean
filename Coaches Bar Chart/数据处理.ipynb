```python
import pandas as pd
df = pd.read_excel('Coaches.xlsx')
df
```

```python
df = (
    df.loc[:, ['Name', 'Discipline']]
    .groupby('Discipline', as_index=False)
    .count()
    .sort_values('Name', ascending=False)
    .rename({'Name': 'Count'}, axis=1)
)
df
```

```python
discipline = df.Discipline.to_list()
count = df.Count.to_list()
```

```python
val = []
for d, c in zip(discipline,count):
      val.append({"discipline": d, "count": c})
```

```python
import json
with open('数据.json', 'w') as f:
    f.write(json.dumps(val))
```
