```python
import pandas as pd
df = pd.read_excel('Medals.xlsx')
df
```

```python
df = (
    df
    .sort_values('Total', ascending=False).head(10)
    .loc[:,['Team/NOC','Gold','Silver','Bronze']]
)
df
```

```python
team = df['Team/NOC'].to_list()
medal = df.iloc[:,[1,2,3]].apply(list,axis=1).to_list()
medal
# glod = df['Gold'].to_list()
# silver = df['Silver'].to_list()
# bronze = df['Bronze'].to_list()
```

```python
def func(i):
    if i==0:
        return 'Gold'
    elif i ==1:
        return 'Silver'
    elif i == 2:
        return 'Bronze'
```

```python
val = []
for t, m in zip(team, medal):
    for i in range(3):
        v = {
            "variety": t,
            "yield": m[i],
            "site": func(i)
        },
        val.append(v)
        # print(v)
# break
    # val = {
    #     'yield': glod,
    #     'variety': team,
    #     'site':
    # }
```

```python
import json
with open ('demo.json') as f:
    print(json.dumps(val))
```
